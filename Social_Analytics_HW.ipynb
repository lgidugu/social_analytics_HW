```python
target_term = ["BBC", "CNN","CBS", "FOX", "NEW YORK Times"] 
sentiment_array = []
for term in target_term:
    counter = 1    
    for x in range(10):
        public_tweets = api.search(term, count=10, result_type="recent")
        for tweet in public_tweets['statuses']:
            tweet_text = tweet['text']         
            compound = analyzer.polarity_scores(tweet['text'])["compound"]
            pos = analyzer.polarity_scores(tweet['text'])["pos"]
            neu = analyzer.polarity_scores(tweet['text'])["neu"]
            neg = analyzer.polarity_scores(tweet['text'])["neg"]            
            tweet_ago = counter
                   
            sentiment_array.append({"media_name": term, "tweet_text":tweet_text, "compound": compound,"positive": pos,"negative": neu,"neutral": neg})
            counter = counter+1  
sentiments_pd  = pd.DataFrame.from_dict(sentiment_array)  
sentiments_pd 
        
```
